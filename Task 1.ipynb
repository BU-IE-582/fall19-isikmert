```python
#********Task 1*********

import numpy as np
import pandas as pd
import csv
from matplotlib import pyplot as plt
from scipy.stats import poisson
from statistics import mean 
```

```python
matches=pd.read_csv('C:\\Users\\MERT\\Desktop\\Dersler\\Data Mining\\Veriler\\matches.csv')
```

```python
my_data = matches[matches.league_id == 148] #filtering only premier league
```

```python
Home_Score = my_data['match_hometeam_score'] #adding column
```

```python
Away_Score = my_data['match_awayteam_score'] #adding column
```

```python
Difference = Home_Score-Away_Score 
```

```python
plt.hist(Home_Score, bins=np.arange(min(Home_Score), max(Home_Score)+1, 1))
plt.title('Home Score Histogram')
plt.xlabel('Number of Games')
plt.ylabel('Home Goals')
avg = mean(Home_Score)
x = np.arange(min(Home_Score),max(Home_Score),1)
plt.plot(x, len(my_data)*poisson.pmf(x,avg),color='black')
plt.show()
```

```python
#poisson distribution is fitted to Home Score Histogram
```

```python
plt.hist(Away_Score, bins=np.arange(min(Away_Score), max(Away_Score)+1, 1), color ="red")
plt.title('Away Score Histogram')
plt.xlabel('Number of Games')
plt.ylabel('Away Goals')
avg = mean(Away_Score)
x = np.arange(min(Away_Score),max(Away_Score),1)
plt.plot(x, len(my_data)*poisson.pmf(x,avg),color='black')
plt.show()
```

```python
#poisson distribution is fitted to Away Score Histogram
```

```python
plt.hist(Difference, bins=np.arange(min(Home_Score-Away_Score), max(Home_Score-Away_Score)+1, 1), color ="green")
plt.title('Home Goals - Away Goals Histogram')
plt.xlabel('Number of Games')
plt.ylabel('Home Goals - Away Goals')
```
